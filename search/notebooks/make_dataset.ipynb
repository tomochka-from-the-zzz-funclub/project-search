```python
import pandas as pd
```

```python
df_movies = pd.read_csv('../data/movie.csv')
df_movies_extra = pd.read_csv('../data/movies_extra.csv')
```

```python
movies = pd.read_csv('../data/movie.csv')
links = pd.read_csv('../data/link.csv')

# Объединение датасетов по колонке movieId
movies_with_tmdb = pd.merge(movies, links[['movieId', 'tmdbId']], on='movieId', how='left')

# Сохранение результата
movies_with_tmdb.to_csv('movies_with_tmdb.csv', index=False)
```

```python
movies_with_tmdb = movies_with_tmdb.dropna(how='any')
movies_with_tmdb['tmdbId'] = movies_with_tmdb['tmdbId'].apply(int)
```

```python
df_movies_extra_cleaned = df_movies_extra.dropna(how='any')
```

```python
df_movies_extra_cleaned['tmdbid'] = df_movies_extra_cleaned['tmdbid'].apply(int)
```

```python
df_movies_extra_cleaned.drop('release_date', axis = 1, inplace = True)
df_movies_extra_cleaned['vote_count'] = df_movies_extra_cleaned['vote_count'].apply(int)
```

```python
movies_with_tmdb = movies_with_tmdb[movies_with_tmdb['genres'] != '(no genres listed)']
movies_with_tmdb
```

```python
df_movies_extra_cleaned.to_csv('df_movies_extra_cleaned.csv', index=False)
```

```python
movies_with_tmdb.to_csv('movies_with_tmdb.csv', index=False)
```

```python
movies_with_tmdb.rename(columns={'tmdbId': 'tmdbid'}, inplace=True)
```

```python
movies_with_tmdb.to_csv('movies_with_tmdb.csv', index=False)
```

```python
merged_movies = pd.merge(movies_with_tmdb, df_movies_extra_cleaned, on='tmdbid', how='inner')

# Сохранение результата (опционально)
# merged_movies.to_csv('merged_movies.csv', index=False)
```

```python
merged_movies['runtime'] = merged_movies['runtime'].apply(int)
```

```python
merged_movies['production_countries'] = merged_movies['production_countries'].str.replace(r'[{}"]', '', regex=True)
merged_movies['spoken_languages'] = merged_movies['spoken_languages'].str.replace(r'[{}]', '', regex=True)
```

```python
# Добавление колонки 'movieId' к merged_movies по совпадению 'tmdbid'
merged_with_movieId = pd.merge(merged_movies, movies_with_tmdb[['movieId', 'tmdbid']], on='tmdbid', how='left')

merged_with_movieId.drop('movieId_y', axis=1, inplace=True)
```

```python
merged_with_movieId.rename(columns={'movieId_x': 'movieId'}, inplace=True)
```

```python
merged_with_movieId['genres'] = merged_with_movieId['genres'].str.replace('|', ',')
```
